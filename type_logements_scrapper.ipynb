## On scrappe ici une page comportant des information sur Résidences principales selon le type de logement et le statut d'occupation :

```python

import bs4
import pandas as pd
from urllib import request
```

## Lecture de la page :

```python
pauvrete = "https://www.insee.fr/fr/statistiques/2012695#tableau-TCRD_003_tab1_departements"

request_text = request.urlopen(pauvrete).read()

print(request_text[:1000])
print("\n")

page = bs4.BeautifulSoup(request_text, "lxml")
page.encode('utf-8')
print(page.find("title"))
```

## Trouver la table :

```python
tableau = page.find('table', {'id': 'produit-tableau-TCRD_003_tab1_departements'})
tableau
```

## Extraction du contenu de la table:

```python
dictionnaire_type_log = {}

en_tete = tableau.find('thead')
for s in en_tete.find_all('th'):
    if len(s)!=0:
        dictionnaire_type_log[s.get_text(strip=True)]=[]
    else : 
        dictionnaire_type_log["Departement"]=[]

dictionnaire_type_log
```

```python
body =  tableau.find('tbody').find_all('tr')

cles = list(dictionnaire_type_log.keys())
print(cles)
nombre_de_colonne =len(cles)

for l in body :
    departement = l.find('th').get_text(strip=True)
    print("departement : ",departement)
    dictionnaire_type_log['Departement'].append(departement)
    trimetres = l.find_all('td')
    print("Trimestres  : ",trimetres)

    for s in range(1,nombre_de_colonne,1) :
        dictionnaire_type_log[cles[s]].append(trimetres[s-1].get_text(strip=True))

print(dictionnaire_type_log)
type_log_df = pd.DataFrame(dictionnaire_type_log)
```
