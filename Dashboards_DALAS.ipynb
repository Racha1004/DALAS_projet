```python
import pandas as pd
```

```python
df = pd.read_csv('Scrapped_Data/Data_APEC_all/all_data_V1_cleaned.csv')
```

```python
df.isna().sum()
```

```python

### Remplacer les valeurs NaN de l'attribut "salaire" avec la moyenne correspondante pour chaque département
moyenne_salaire_par_departement = df.groupby("departement")["salaire"].mean().reset_index()
moyenne_salaire_par_departement.isna().sum()
### Remplacer les valeurs NaN par la moyenne générale des salaires
moyenne_generale_salaire = df["salaire"].mean()
moyenne_salaire_par_departement["salaire"].fillna(moyenne_generale_salaire, inplace=True)
moyenne_salaire_par_departement

for dep in moyenne_salaire_par_departement["departement"].unique() :
    mask = df["departement"] == dep
    new_salaire = moyenne_salaire_par_departement.loc[moyenne_salaire_par_departement["departement"] == dep, "salaire"].iloc[0]
    df.loc[mask, "salaire"] = df.loc[mask, "salaire"].fillna(new_salaire)

df.isna().sum()
### Remplacer les valeurs NaN de l'attribut "Teletravail" avec la categorie "non specifié":
df['Teletravail'] = df['Teletravail'].fillna('non spécifié')
df.head(10)
```

```python
df.columns
```

```python
# Définir une fonction pour remplacer les valeurs
liste_valeurs_autorisees= ['CDI', 'CDD', "Mission d'intérim"]
def remplacer_par(valeur):
    if valeur in liste_valeurs_autorisees:
        return valeur
    else:
        return 'Alternance'

# Appliquer la fonction à la colonne
df['type_contrat'] = df['type_contrat'].apply(remplacer_par)

df['type_contrat'].unique()
```

```python
df_filtred = df[[ "entreprise",'type_contrat',
       'salaire', 'experience', 'metier',
       'zone_de_deplacement',  'Teletravail',
       'departement','statut_du_poste']]
df_filtred
```

## Filtrer par domaine :

```python
df_filtred['metier'].unique()
```

```python
dev=['Développeur','Architecte développement', 'Ingénieur test et recettes','Ingénieur test et essai', 'Ingénieur en études et développement informatiques',
       'Ingénieur de maintenance informatique', 'Ingénieur intégration progiciel', 'Consultant technique progiciel','Business developer'     ]

data = ['Data engineer','Ingénieur R&D','Administrateur de bases de données','Analyste financier',
        'Chef de projet R&D', 'Data analyst','Chief Data Officer', 'Data scientist','Administrateur data center',
         'Architecte Big Data', 'Délégué à la protection des données','Web analyst' ]



reseaux = [  'Ingénieur télécoms', 'Ingénieur process','Architecte infrastructures',
           'Administrateur réseaux','Ingénieur réseaux', 'Administrateur système','Ingénieur système',
              'Architecte réseaux',   'Ingénieur traitement du signal','Architecte système']

securite = ['Ingénieur sécurité informatique', 'Responsable sécurité informatique','Consultant en cybersécurité',]

miage = ['Responsable informatique','Ingénieur informatique industrielle',
         'Ingénieur de production informatique', 'Project management officer',
         'Ingénieur calcul et structure',
          'Chef de projet informatique','Chef de projet Agile',
         'Chef de projet travaux','Ingénieur développement BI',
         'Business manager','Business analyst',       'Chef de projet informatique décisionnelle',
         'Directeur de projets informatiques',
         'Chef de projet digital','Chef de projet fonctionnel web' ]
```

```python
for dom in [dev,data,reseaux,securite,miage]:
  metiers_inclus = df_filtred[df_filtred['metier'].isin(dom)]

  # Compter le nombre de ces lignes
  nombre_lignes = len(metiers_inclus)
  print(nombre_lignes)
```

```python
# Fonction pour attribuer le domaine
def assign_domaine(metier):
    if metier in dev:
        return 'développement'
    elif metier in data:
        return 'data'
    elif metier in reseaux:
        return 'réseaux'
    elif metier in securite:
        return 'securité'
    elif metier in miage:
        return 'miage'
    else:
        return 'autre'  # Pour les métiers qui ne sont dans aucune liste

# Appliquer la fonction pour créer la colonne 'domaine'
df_filtred['domaine'] = df_filtred['metier'].apply(assign_domaine)
```

```python
df_filtred
```

```python
# Grouper par département et domaine et compter les offres
df_filtred_groupedBy_dep = df_filtred.groupby(['departement', 'domaine']).size().unstack(fill_value=0)
# df_filtred_groupedBy_dep.columns =[["deparetement",	"data",	"développement",	"miage",	"réseaux",	"securité",	"autre"]]

df_filtred_groupedBy_dep.columns
```

```python
df_filtred_groupedBy_dep_reset = df_filtred_groupedBy_dep.reset_index()

df_filtred_groupedBy_dep_reset.columns
```

## Plusieurs application :

```python
from dash import Dash, html, dash_table, dcc, callback, Output, Input
import pandas as pd
import plotly.express as px
import dash_bootstrap_components as dbc

external_stylesheets = [dbc.themes.CERULEAN]
app = Dash(__name__, external_stylesheets=external_stylesheets)

liste_departements = df_filtred['departement'].unique()

options_departements = [{'label': departement, 'value': departement} for departement in liste_departements]

liste_domaines = df_filtred['domaine'].unique()

# App layout
layout_pie_nb_offers_per_departemt = dbc.Container([
    dbc.Row([
        html.Div("Nombre d'offres par domaine pour un département sélectionné", className="text-primary text-center fs-3")
    ]),

    dbc.Row([
        dbc.Col([
            dcc.Dropdown(
                options=options_departements,
                value=liste_departements[0],  # Valeur par défaut
                id='dropdown-departement1',
                clearable=False
            )
        ])
    ]),

    dbc.Row([
        dbc.Col([
            dcc.Graph(id='pie-chart-offres')
        ])
    ]),
], fluid=True)

# Add controls to build the interaction
@app.callback(
    Output('pie-chart-offres', 'figure'),
    [Input('dropdown-departement1', 'value')]
)
def update_pie_chart(selected_departement):
    # Filtrer le DataFrame pour le département sélectionné
    df_selected_departement = df_filtred[df_filtred['departement'] == selected_departement]

    # Compter le nombre d'offres pour chaque domaine
    df_domaine_count = df_selected_departement['domaine'].value_counts().reset_index()
    df_domaine_count.columns = ['domaine', 'nombre_offres']

    # Créer le camembert
    fig = px.pie(df_domaine_count, values='nombre_offres', names='domaine', title=f"Nombre d'offres par domaine dans le département '{selected_departement}'")

    return fig
layout_offres_par_departement = dbc.Container([
    dbc.Row([
        html.Div('Nombre d\'offres par département pour un domaine donné', className="text-primary text-center fs-3")
    ]),

    dbc.Row([
        dbc.RadioItems(options=[{"label": x, "value": x} for x in ['autre', 'data', 'développement', 'miage', 'réseaux', 'securité']],
                       value='data',
                       inline=True,
                       id='radio-buttons-final')
    ]),

    dbc.Row([
        dbc.Col([
            dash_table.DataTable(data=df_filtred_groupedBy_dep_reset.to_dict('records'), page_size=12, style_table={'overflowX': 'auto'})
        ], width=6),

        dbc.Col([
            dcc.Graph(figure={}, id='my-first-graph-final')
        ], width=6),
    ]),

], fluid=True)

# Add controls to build the interaction for the first part
@app.callback(
    Output(component_id='my-first-graph-final', component_property='figure'),
    Input(component_id='radio-buttons-final', component_property='value')
)
def update_graph(col_chosen):
    fig = px.histogram(df_filtred_groupedBy_dep_reset, x='departement', y=col_chosen, histfunc='avg')
    fig.update_yaxes(title_text="Nombre d'offres")

    return fig

## App layout
layout_nb_entreprise_par_domaine = dbc.Container([
    dbc.Row([
        html.Div("Nombre d'entreprises proposant des postes dans un domaine sélectionné (par département)", className="text-primary text-center fs-3")
    ]),

    dbc.Row([
        dbc.RadioItems(
            options=[{"label": x, "value": x} for x in ['autre', 'data', 'développement', 'miage', 'réseaux', 'sécurité']],
            value='data',
            inline=True,
            id='radio-buttons-final2'
        )
    ]),

    dbc.Row([
        dbc.Col([
            dash_table.DataTable(
                data=df_filtred_groupedBy_dep_reset.to_dict('records'),
                page_size=12,
                style_table={'overflowX': 'auto'}
            )
        ], width=6),

        dbc.Col([
            dcc.Graph(figure={}, id='my-second-graph-final')
        ], width=6),
    ]),

], fluid=True)

# Add controls to build the interaction
@app.callback(
    Output(component_id='my-second-graph-final', component_property='figure'),
    Input(component_id='radio-buttons-final2', component_property='value')
)
def update_graph(selected_job):
    df_selected_job = df_filtred[df_filtred['domaine'] == selected_job]

    df_grouped = df_selected_job.groupby('departement')['entreprise'].nunique().reset_index()

    # Créer le graphique à barres
    fig = px.bar(df_grouped, x='departement', y='entreprise', color='departement', labels={'entreprise': 'Nombre d\'entreprises', 'departement': 'Département'})

    fig.update_layout(title_text=f"Nombre d'entreprises proposant des postes dans le domaine de '{selected_job}' par département")

    return fig

liste_metiers = df_filtred['metier'].unique()

options_metiers = [{'label': metier, 'value': metier} for metier in liste_metiers]

layer_entreprise = dbc.Container([
    dbc.Row([
        html.Div("Fréquence des entreprises proposant un métier sélectionné par département", className="text-primary text-center fs-3")
    ]),

    dbc.Row([
        dcc.Dropdown(
            options=options_metiers,
            value=liste_metiers[0],  # Valeur par défaut
            id='dropdown-entreprise',
            clearable=False
        )
    ]),

    dbc.Row([
        dbc.Col([
            dash_table.DataTable(
                data=df_filtred_groupedBy_dep_reset.to_dict('records'),
                page_size=12,
                style_table={'overflowX': 'auto'}
            )
        ], width=6),

        dbc.Col([
            dcc.Graph(figure={}, id='my-third-graph-final')
        ], width=6),
    ]),

], fluid=True)

@app.callback(
    Output(component_id='my-third-graph-final', component_property='figure'),
    Input(component_id='dropdown-entreprise', component_property='value')
)
def update_graph(selected_job):
    df_selected_job = df_filtred[df_filtred['metier'] == selected_job]

    df_grouped = df_selected_job.groupby('departement')['entreprise'].nunique().reset_index()

    fig = px.bar(df_grouped, x='departement', y='entreprise', color='departement', labels={'entreprise': 'Nombre d\'entreprises', 'departement': 'Département'})

    fig.update_layout(title_text=f"Fréquence des entreprises proposant le métier '{selected_job}' par département")

    return fig


liste_departements = df_filtred['departement'].unique()

options_departements = [{'label': departement, 'value': departement} for departement in liste_departements]

options_graphiques = [
    {'label': 'Histogramme', 'value': 'histogramme'},
    {'label': 'Box Plot', 'value': 'boxplot'},
    {'label': 'Violon', 'value': 'violin'}
]

layer_salary = dbc.Container([
    dbc.Row([
        html.Div("Distribution des salaires par département", className="text-primary text-center fs-3")
    ]),

    dbc.Row([
        dbc.Col([
            dcc.Dropdown(
                options=options_departements,
                value=liste_departements[0],  # Valeur par défaut
                id='dropdown-departement',
                clearable=False
            )
        ]),
        dbc.Col([
            dcc.Dropdown(
                options=options_graphiques,
                value='histogramme',  # Valeur par défaut
                id='dropdown-graphique',
                clearable=False
            )
        ])
    ]),

    dbc.Row([
        dbc.Col([
            dcc.Graph(id='graph-salaire')
        ])
    ]),
], fluid=True)

@app.callback(
    Output('graph-salaire', 'figure'),
    [Input('dropdown-departement', 'value'),
     Input('dropdown-graphique', 'value')]
)
def update_graph(selected_departement, selected_graphique):
    df_selected_departement = df_filtred[df_filtred['departement'] == selected_departement]

    if selected_graphique == 'histogramme':
        fig = px.histogram(df_selected_departement, x='salaire', nbins=20, labels={'salaire': 'Salaire', 'count': 'Nombre de salariés'})
    elif selected_graphique == 'boxplot':
        fig = px.box(df_selected_departement, x='departement', y='salaire', labels={'salaire': 'Salaire', 'departement': 'Département'})
    elif selected_graphique == 'violin':
        fig = px.violin(df_selected_departement, x='departement', y='salaire', labels={'salaire': 'Salaire', 'departement': 'Département'})

    fig.update_layout(title_text=f"Distribution des salaires dans le département '{selected_departement}' ({selected_graphique})")

    return fig

app.layout = html.Div([layout_pie_nb_offers_per_departemt,layout_offres_par_departement, layout_nb_entreprise_par_domaine,layer_entreprise,layer_salary])

# Run the app
if __name__ == '__main__':
    app.run_server(debug=True)
```
