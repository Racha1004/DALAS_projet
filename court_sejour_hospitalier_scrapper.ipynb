## On scrappe ici une page comportant des information sur le court séjour hospitalier  :

```python

import bs4
import pandas as pd
from urllib import request
```

## Lecture de la page :

```python
pauvrete = "https://www.insee.fr/fr/statistiques/2012675#graphique-TCRD_066_tab1_regions2016"

request_text = request.urlopen(pauvrete).read()

print(request_text[:1000])
print("\n")

page = bs4.BeautifulSoup(request_text, "lxml")
page.encode('utf-8')
print(page.find("title"))
```

## Trouver la table :

```python
tableau = page.find('table', {'id': 'produit-tableau-TCRD_066_tab1_departements'})
tableau
```

## Extraction du contenu de la table:

```python
dictionnaire_cs_hosp = {}

en_tete = tableau.find('thead')
en_tete_groupe = en_tete.find_all('tr')
colonnes= []
for groupe in en_tete_groupe:
    for elem in groupe.find_all('th'):
        colonnes.append(elem.get_text(strip=True))


dictionnaire_cs_hosp["Departement"]=[]
dictionnaire_cs_hosp["["+colonnes[1]+"]"+colonnes[5]]=[]
dictionnaire_cs_hosp["["+colonnes[1]+"]"+colonnes[6]]=[]
dictionnaire_cs_hosp["["+colonnes[1]+"]"+colonnes[7]]=[]

dictionnaire_cs_hosp["["+colonnes[2]+"]"+colonnes[8]]=[]
dictionnaire_cs_hosp["["+colonnes[2]+"]"+colonnes[9]]=[]

dictionnaire_cs_hosp["["+colonnes[3]+"]"+colonnes[10]]=[]

dictionnaire_cs_hosp
```

```python
body =  tableau.find('tbody').find_all('tr')

cles = list(dictionnaire_cs_hosp.keys())
print(cles)
nombre_de_colonne =len(cles)

for l in body :
    departement = l.find('th').get_text(strip=True)
    print("departement : ",departement)
    dictionnaire_cs_hosp['Departement'].append(departement)
    trimetres = l.find_all('td')
    print("Trimestres  : ",trimetres)

    for s in range(1,nombre_de_colonne,1) :
        dictionnaire_cs_hosp[cles[s]].append(trimetres[s-1].get_text(strip=True))

print(dictionnaire_cs_hosp)
dictionnaire_cs_hosp_df = pd.DataFrame(dictionnaire_cs_hosp)
```
