# Importation des bibliothéques

```python
import bs4
import lxml
import pandas as pd
import urllib

from urllib import request
```

Ouverture du lien afin de scrapper la population de chaque département en France

```python
URL = "https://france.ousuisje.com/departements/classement/superficie.php"

request_text = request.urlopen(URL).read()
page = bs4.BeautifulSoup(request_text, "lxml")
print(page.find("title"))
```

Récupération de la table 

```python
table = page.find('table')

#récupération des colonnes et création d'un dictionnaire
columns = table.find_all('th')
data_Pop = {}
for col in columns :
    data_Pop[col.get_text(strip=True)] = []
print()

lines = table.find_all('tr')[1:]
for l in lines:
    values = l.find_all('td')
    for i, col in enumerate(data_Pop.keys()):
        data_Pop[col].append(values[i].get_text(strip=True))

data_Pop
```

# Créer un DataFrame à partir des données

```python
df = pd.DataFrame(data_Pop)
df
```

# Enregistrer le DataFrame au format CSV

```python
df.to_csv('./Data/data_Pop.csv', index=False)
```
