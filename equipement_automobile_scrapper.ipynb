## On scrappe ici une page comportant le taux de pauvreté selon l'âge  :

```python

import bs4
import pandas as pd
from urllib import request
```

## Lecture de la page :

```python
pauvrete = "https://www.insee.fr/fr/statistiques/2012803#tableau-TCRD_024_tab1_departements"

request_text = request.urlopen(pauvrete).read()

print(request_text[:1000])
print("\n")

page = bs4.BeautifulSoup(request_text, "lxml")
page.encode('utf-8')
print(page.find("title"))
```

## Trouver la table :

```python
tableau = page.find('table', {'id': 'produit-tableau-TCRD_024_tab1_departements'})
tableau
```

## Extraction du contenu de la table:

```python
dictionnaire_pauvrete = {}

en_tete = tableau.find('thead')
for s in en_tete.find_all('th'):
    if len(s)!=0:
        dictionnaire_pauvrete[s.get_text(strip=True)]=[]
    else : 
        dictionnaire_pauvrete["Departement"]=[]

dictionnaire_pauvrete
```

```python
body =  tableau.find('tbody').find_all('tr')

cles = list(dictionnaire_pauvrete.keys())
print(cles)
nombre_de_colonne =len(cles)

for l in body :
    departement = l.find('th').get_text(strip=True)
    print("departement : ",departement)
    dictionnaire_pauvrete['Departement'].append(departement)
    trimetres = l.find_all('td')
    print("Trimestres  : ",trimetres)

    for s in range(1,nombre_de_colonne,1) :
        dictionnaire_pauvrete[cles[s]].append(trimetres[s-1].get_text(strip=True))

print(dictionnaire_pauvrete)
pauvrete_df = pd.DataFrame(dictionnaire_pauvrete)
```
