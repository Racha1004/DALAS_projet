## On scrappe ici une page comportant des information sur l'equipement automobile dans differents appartement  :

```python

import bs4
import pandas as pd
from urllib import request
```

## Lecture de la page :

```python
pauvrete = "https://www.insee.fr/fr/statistiques/2012694#tableau-TCRD_001_tab1_departements"

request_text = request.urlopen(pauvrete).read()

print(request_text[:1000])
print("\n")

page = bs4.BeautifulSoup(request_text, "lxml")
page.encode('utf-8')
print(page.find("title"))
```

## Trouver la table :

```python
tableau = page.find('table', {'id': 'produit-tableau-TCRD_001_tab1_departements'})
tableau
```

## Extraction du contenu de la table:

```python
dictionnaire_equip_auto = {}

en_tete = tableau.find('thead')
for s in en_tete.find_all('th'):
    if len(s)!=0:
        dictionnaire_equip_auto[s.get_text(strip=True)]=[]
    else : 
        dictionnaire_equip_auto["Departement"]=[]

dictionnaire_equip_auto
```

```python
body =  tableau.find('tbody').find_all('tr')

cles = list(dictionnaire_equip_auto.keys())
print(cles)
nombre_de_colonne =len(cles)

for l in body :
    departement = l.find('th').get_text(strip=True)
    print("departement : ",departement)
    dictionnaire_equip_auto['Departement'].append(departement)
    trimetres = l.find_all('td')
    print("Trimestres  : ",trimetres)

    for s in range(1,nombre_de_colonne,1) :
        dictionnaire_equip_auto[cles[s]].append(trimetres[s-1].get_text(strip=True))

print(dictionnaire_equip_auto)
equip_auto_df = pd.DataFrame(dictionnaire_equip_auto)
```
